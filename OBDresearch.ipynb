Импортируем библиотеки и данные:

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("torqueTrackLog.csv")
```

Данные содержат разделительную линию где-то в файле в виде заголовков. Найдем индекс этой линии и разделим набор данных

```python
split = df[df['GPS Time']=='GPS Time'].index[0]
df1 = df.iloc[:split]
```

Преобразуем время измерений в индекс датасета

```python
time = pd.to_datetime(df1['GPS Time'])
```

```python
df1.index = time
```
